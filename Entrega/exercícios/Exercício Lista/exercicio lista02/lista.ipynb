# Listas em Python 🐍

## O que é uma Lista?

Uma **lista (list)** é uma coleção ordenada de elementos que pode armazenar diferentes tipos de dados.

As listas são uma das estruturas mais utilizadas em Python porque permitem:

* Armazenar vários valores em uma única variável;
* Adicionar novos elementos;
* Remover elementos;
* Alterar valores;
* Percorrer os dados facilmente.

As listas são criadas utilizando **colchetes `[]`**.

---

# Características das Listas

✔ Mantêm a ordem dos elementos

✔ Permitem valores repetidos

✔ Permitem diferentes tipos de dados

✔ São mutáveis (podem ser alteradas)

✔ Possuem índices

---

# Criando uma Lista

```python
frutas = ["maçã", "banana", "uva"]

print(frutas)
```

Saída:

```python
['maçã', 'banana', 'uva']
```

---

# Lista com Diferentes Tipos

```python
dados = ["Ana", 20, 8.5, True]

print(dados)
```

Saída:

```python
['Ana', 20, 8.5, True]
```

---

# Acessando Elementos

Cada elemento possui uma posição (índice).

```python
frutas = ["maçã", "banana", "uva"]

print(frutas[0])
print(frutas[1])
print(frutas[2])
```

Saída:

```python
maçã
banana
uva
```

---

# Índices Negativos

```python
frutas = ["maçã", "banana", "uva"]

print(frutas[-1])
print(frutas[-2])
```

Saída:

```python
uva
banana
```

---

# Alterando um Elemento

```python
frutas = ["maçã", "banana", "uva"]

frutas[1] = "laranja"

print(frutas)
```

Saída:

```python
['maçã', 'laranja', 'uva']
```

---

# Adicionando Elementos

## append()

Adiciona ao final da lista.

```python
frutas = ["maçã", "banana"]

frutas.append("uva")

print(frutas)
```

Saída:

```python
['maçã', 'banana', 'uva']
```

---

## insert()

Adiciona em uma posição específica.

```python
frutas = ["maçã", "uva"]

frutas.insert(1, "banana")

print(frutas)
```

Saída:

```python
['maçã', 'banana', 'uva']
```

---

# Removendo Elementos

## remove()

Remove pelo valor.

```python
frutas = ["maçã", "banana", "uva"]

frutas.remove("banana")

print(frutas)
```

Saída:

```python
['maçã', 'uva']
```

---

## pop()

Remove pela posição.

```python
frutas = ["maçã", "banana", "uva"]

frutas.pop(1)

print(frutas)
```

Saída:

```python
['maçã', 'uva']
```

---

## del

```python
frutas = ["maçã", "banana", "uva"]

del frutas[0]

print(frutas)
```

Saída:

```python
['banana', 'uva']
```

---

# Tamanho da Lista

```python
frutas = ["maçã", "banana", "uva"]

print(len(frutas))
```

Saída:

```python
3
```

---

# Verificando se um Elemento Existe

```python
frutas = ["maçã", "banana", "uva"]

print("banana" in frutas)
print("laranja" in frutas)
```

Saída:

```python
True
False
```

---

# Percorrendo uma Lista

```python
frutas = ["maçã", "banana", "uva"]

for fruta in frutas:
    print(fruta)
```

Saída:

```python
maçã
banana
uva
```

---

# Percorrendo com Índices

```python
frutas = ["maçã", "banana", "uva"]

for i in range(len(frutas)):
    print(i, "-", frutas[i])
```

Saída:

```python
0 - maçã
1 - banana
2 - uva
```

---

# Fatiamento (Slice)

```python
numeros = [10, 20, 30, 40, 50]

print(numeros[1:4])
```

Saída:

```python
[20, 30, 40]
```

---

# Ordenando uma Lista

```python
numeros = [5, 2, 8, 1]

numeros.sort()

print(numeros)
```

Saída:

```python
[1, 2, 5, 8]
```

---

# Ordem Decrescente

```python
numeros = [5, 2, 8, 1]

numeros.sort(reverse=True)

print(numeros)
```

Saída:

```python
[8, 5, 2, 1]
```

---

# Lista de Números

```python
numeros = [10, 20, 30, 40, 50]

print("Maior:", max(numeros))
print("Menor:", min(numeros))
print("Soma:", sum(numeros))
```

Saída:

```python
Maior: 50
Menor: 10
Soma: 150
```

---

# Lista Dentro de Lista

```python
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matriz[1][2])
```

Saída:

```python
6
```

---

# Exemplo Prático – Notas dos Alunos

```python
notas = [8.5, 7.0, 9.0, 10]

media = sum(notas) / len(notas)

print("Média:", media)
```

Saída:

```python
Média: 8.625
```

---

# Exemplo Prático – Cadastro de Nomes

```python
nomes = []

for i in range(3):
    nome = input("Digite um nome: ")
    nomes.append(nome)

print(nomes)
```

---

# Exemplo Prático – Lista de Compras

```python
compras = []

while True:
    item = input("Digite um item (fim para sair): ")

    if item.lower() == "fim":
        break

    compras.append(item)

print("Lista de compras:")
print(compras)
```

# Comparação entre Coleções

| Característica    | Lista | Tupla | Set | Dicionário |
| ----------------- | ----- | ----- | --- | ---------- |
| Ordenada          | ✅     | ✅     | ❌   | ✅          |
| Permite repetição | ✅     | ✅     | ❌   | Chaves não |
| Mutável           | ✅     | ❌     | ✅   | ✅          |
| Índices           | ✅     | ✅     | ❌   | ❌          |
| Chave e valor     | ❌     | ❌     | ❌   | ✅          |

---

# Resumo

| Operação             | Comando    |
| -------------------- | ---------- |
| Criar lista          | `[]`       |
| Adicionar            | `append()` |
| Inserir              | `insert()` |
| Remover              | `remove()` |
| Remover por índice   | `pop()`    |
| Tamanho              | `len()`    |
| Ordenar              | `sort()`   |
| Somar valores        | `sum()`    |
| Maior valor          | `max()`    |
| Menor valor          | `min()`    |
| Verificar existência | `in`       |

### Quando usar Listas?

✅ Cadastro de alunos

✅ Notas de uma turma

✅ Lista de compras

✅ Produtos de um sistema

✅ Dados coletados do usuário

✅ Qualquer situação onde os dados precisem ser alterados ao longo da execução do programa.
